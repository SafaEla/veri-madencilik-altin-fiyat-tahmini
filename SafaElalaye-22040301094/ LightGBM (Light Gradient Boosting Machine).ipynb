```python
# Gerekli Kütüphaneler
import pandas as pd # Veri analizi için pandas kütüphanesini içe aktarıyoruz. Excel veya CSV dosyalarını okumakta çok işe yarıyor.
import numpy as np  # Matematiksel işlemler ve dizilerle çalışmak için numpy kütüphanesini ekliyoruz.
import seaborn as sns    # Daha estetik grafikler için seaborn kütüphanesi ekleniyor
import matplotlib.pyplot as plt   # Grafik çizimleri için matplotlib kütüphanesi kullanılıyor
import lightgbm as lgb  # Kütüphane değişti
from sklearn.model_selection import TimeSeriesSplit, RandomizedSearchCV  # Zaman serisi verileri için özel olarak kullanılan çapraz doğrulama yöntemi
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score   # Model doğruluğunu ölçmek için MSE, MAE ve R² performans metrikleri
from sklearn.model_selection import train_test_split   # Veriyi eğitim ve test seti olarak rastgele bölen fonksiyon
from sklearn.preprocessing import StandardScaler     # Özellikleri standartlaştırmak için kullanılan ölçekleyici
```

```python
df = pd.read_csv(r"XAU_15m_data_2004_to_2024-20-09.csv")   # CSV dosyasından veriyi okuyoruz

print("\nVerinin boyutu (satır × sütun):")
print(df.shape)   # Kaç satır ve sütun olduğunu gösteriyor

print("\nVerinin ilk 5 satırı:")
display(df.head())  # İlk 5 satıra bakıyoruz

if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')  # Tarih sütununu datetime tipine çevirme

df = df.sort_values(by="Date")   # Veriyi tarih sırasına göre sıralama

print("\nEksik değer sayısı:")
display(df.isnull().sum())   # Hangi sütunda kaç eksik var bakıyoruz
```

```python
print("\nTemizlik öncesi eksik değerler:")
print(df.isnull().sum())    # Çoğu altın verisinde eksik değer yok, ama yine de kontrol edelim:

numeric_cols = ['Open', 'High', 'Low', 'Close', 'Volume']  # Sayısal sütunlar

for col in numeric_cols:
    if df[col].isnull().sum() > 0:
        df[col].fillna(df[col].mean(), inplace=True)   # Eksikleri ortalama ile dolduruyorum

if 'Date' in df.columns:  # Tarih sütununda eksik varsa = satırı sil
    df = df.dropna(subset=['Date'])  

print("\nEksik değerler işlendi.")   # Temizlik bitti

numeric_cols = ['Open', 'High', 'Low', 'Close', 'Volume']   # Tekrar kontrol için liste

for col in numeric_cols:
    Q1 = df[col].quantile(0.25)  # İlk çeyrek
    Q3 = df[col].quantile(0.75)  # Üçüncü çeyrek
    IQR = Q3 - Q1                # IQR hesaplama
    
    lower_bound = Q1 - 1.5 * IQR  # Alt sınır
    upper_bound = Q3 + 1.5 * IQR  # Üst sınır
    
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]   # Aykırı değerleri buluyor
    print(f"{col}: Aykırı değer sayısı = {len(outliers)}")            # Kaç tane olduklarını yazdırıyor
```

```python
# Feature engineering işlemleri yapılıyor

if 'Time' in df.columns:   # Eğer Time sütunu varsa Date ile birleştirilerek tam zaman oluşturuluyor
    df['Datetime'] = pd.to_datetime(df['Date'].dt.strftime('%Y-%m-%d') + ' ' + df['Time'])
else:     # Yoksa sadece Date üzerinden Datetime oluşturuluyor
    df['Datetime'] = df['Date']

df = df.sort_values('Datetime').reset_index(drop=True)    # Veriyi zaman sırasına göre sıralama

df = df[df['Close'].diff() != 0].reset_index(drop=True)   # Önceki kapanış fiyatı ile aynıysa tekrar eden satırları kaldırma

# Zaman özellikleri ekleme
df['Year'] = df['Datetime'].dt.year
df['Month'] = df['Datetime'].dt.month
df['Day'] = df['Datetime'].dt.day
df['Hour'] = df['Datetime'].dt.hour
df['Weekday'] = df['Datetime'].dt.weekday  # 0 = Pazartesi, 6 = Pazar

# Temel fiyat farkı özellikleri hesaplanıyor
df['High_Low_Diff'] = df['High'] - df['Low']
df['Close_Open_Diff'] = df['Close'] - df['Open']
df['High_Close_Diff'] = df['High'] - df['Close']
df['Low_Close_Diff'] = df['Low'] - df['Close']

# Hareketli ortalama ve standart sapma özellikleri oluşturuluyor
for w in [3,5,10]:
    df[f'MA_Close_{w}'] = df['Close'].rolling(window=w).mean()
    df[f'STD_Close_{w}'] = df['Close'].rolling(window=w).std()
    df[f'MA_High_{w}'] = df['High'].rolling(window=w).mean()
    df[f'MA_Low_{w}'] = df['Low'].rolling(window=w).mean()

# Fiyatın önceki kapanış ile farkı (lag features)
df['Close_Lag1'] = df['Close'].shift(1)
df['Close_Lag2'] = df['Close'].shift(2)
df['Close_Lag3'] = df['Close'].shift(3)

# Momentum hesaplamaları yapılıyor
df['Momentum_1'] = df['Close'] - df['Close_Lag1']
df['Momentum_2'] = df['Close'] - df['Close_Lag2']

# İlk 10 satırı göster
print("\nGeliştirilmiş Feature Engineering sonrası veri:")
display(df.head(10))
```

```python
# veriyi %80 eğitim / %20 test olarak bölüm
df.dropna(inplace=True)   # Veri içindeki eksik kayıtlar temizleniyor

y = df['Close']     # Tahmin edilecek hedef değer olarak 'Close' sütunu seçiliyor

# Modelde kullanılacak özellik sütunları belirleniyor
X = df.drop(['Close', 'Date', 'Datetime', 'Time'], axis=1, errors='ignore')   


# Veri %80 eğitim, %20 test olarak ayrılıyor
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, shuffle=False, random_state=42
)

# Özelliklerin ölçeklendirilmesi için standartlaştırıcı tanımlanıyor
scaler = StandardScaler()

# Eğitim verisi ölçeklendiriliyor
X_train_scaled = scaler.fit_transform(X_train)
# Test verisi aynı scaler ile dönüştürülüyor
X_test_scaled = scaler.transform(X_test)

# Son durumu gösterme
print("\nVeriler modeller için hazır:")
print(f"X_train: {X_train_scaled.shape}, X_test: {X_test_scaled.shape}")
print(f"y_train: {y_train.shape}, y_test: {y_test.shape}")
```

```python
df = pd.read_csv(r"XAU_15m_data_2004_to_2024-20-09.csv") 
# Sadece sayısal sütunları alalım
numerical_df = df.select_dtypes(include=[np.number])

# Korelasyon hesapla
corr_matrix = numerical_df.corr()

# Hedef değişken (Close) ile en yüksek korelasyona sahip 10 özelliği bul
top_corr_features = corr_matrix.index[abs(corr_matrix["Close"]) > 0.5]

plt.figure(figsize=(12, 10))
plt.title("Özellikler Arası Korelasyon Matrisi (Heatmap)")
# Sadece en ilişkili özellikleri çizelim ki grafik karışmasın
sns.heatmap(df[top_corr_features].corr(), annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)
plt.show()
```

```python
# VERİ HAZIRLIĞI VE DURAĞANLAŞTIRMA
df_model = df.copy()  # Orijinal veriyi bozmamak için df'nin bir kopyası alınıyor

print("Veri Hazırlanıyor: LightGBM için Delta (Fark) Dönüşümü...")

# Bir sonraki kapanış fiyatı ile mevcut fiyat arasındaki fark hedef değişken olarak oluşturuluyor
df_model['Target_Diff'] = df_model['Close'].shift(-1) - df_model['Close']   

# Zaman pencereleri için gecikmeli fark özellikleri ekleniyor
for i in [1, 2, 3, 5, 8, 12, 24]:
    df_model[f'Diff_Lag_{i}'] = df_model['Close'].diff(i)   # "i" periyot önceki fiyattan fark hesaplanarak özellik oluşturuluyor


delta = df_model['Close'].diff()   # Fiyat değişimi hesaplanıyor
gain = (delta.where(delta > 0, 0)).rolling(window=14).mean()     # Yalnızca pozitif değişimlerin 14 periyotluk ortalaması alınıyor
loss = (-delta.where(delta < 0, 0)).rolling(window=14).mean()    # Yalnızca negatif değişimlerin 14 periyotluk ortalaması alınıyor
rs = gain / loss     # RSI için gerekli güç oranı hesaplanıyor
df_model['RSI'] = 100 - (100 / (1 + rs))   # Fiyat momentumunu ölçen RSI göstergesi hesaplanıp ekleniyor

# Volatilite ve SMA Farkı
df_model['Volatility'] = df_model['Close'].rolling(window=20).std()      # Fiyatın son 20 periyottaki oynaklığı hesaplanarak modele dahil ediliyor
df_model['SMA_Dist'] = df_model['Close'] - df_model['Close'].rolling(window=50).mean()     # Fiyatın 50 dönemlik ortalamadan uzaklığı hesaplanıyor

df_model.dropna(inplace=True) # Oluşan boş değerli satırlar veri bütünlüğü için temizleniyor
```

```python
# VERİ BÖLME 
y = df_model['Target_Diff']    # Tahmin edilmesi gereken hedef değişken seçiliyor
# Modelde kullanılmayacak sütunlar çıkarılarak özellik matrisi oluşturuluyor
X = df_model.drop(['Close', 'Date', 'Time', 'Datetime', 'Open', 'High', 'Low', 'Target_Diff'], axis=1, errors='ignore')

split_point = int(len(X) * 0.8)    # Verinin yüzde 80'i eğitim, yüzde 20'si test olacak şekilde bölme noktası belirleniyor
X_train, X_test = X.iloc[:split_point], X.iloc[split_point:]    # Özellikler eğitim ve test seti olarak ayrılıyor
y_train, y_test = y.iloc[:split_point], y.iloc[split_point:]    # Hedef değişken eğitim ve test setine paylaştırılıyor

original_close_test = df_model['Close'].iloc[split_point:]  # Test dönemindeki gerçek kapanış fiyatları referans için saklanıyor

print(f"Eğitim Verisi: {X_train.shape}, Test Verisi: {X_test.shape}")
```

```python
# LIGHTGBM MODELLEME VE OPTİMİZASYON 
print("\nLightGBM Modeli Optimize Ediliyor...")

lgb_model = lgb.LGBMRegressor(random_state=42, n_jobs=-1, verbose=-1)  # LightGBM regresyon modeli oluşturuluyor, çoklu çekirdek kullanılıyor ve çıktılar kapalı

# LightGBM'e özel parametreler
param_dist = {
    'n_estimators': [500, 1000],    # Kullanılacak ağaç sayısı
    'learning_rate': [0.01, 0.05, 0.1],  # Öğrenme oranları
    'num_leaves': [31, 50, 70],     # Tek bir ağacın maksimum yaprak sayısı
    'min_child_samples': [20, 50],     # Yaprak başına minimum örnek sayısı
    'subsample': [0.8, 0.9],       # Rastgele örnekleme oranı (satırlar)
    'colsample_bytree': [0.8, 0.9]      # Rastgele örnekleme oranı (sütunlar)
}

tscv = TimeSeriesSplit(n_splits=3)    # Zaman serisi verileri için 3 katlı çapraz doğrulama tanımlanıyor

random_search = RandomizedSearchCV(
    estimator=lgb_model,    # Optimizasyon yapılacak model
    param_distributions=param_dist,   # Hiperparametre aralığı
    n_iter=10,     # Rastgele denenecek parametre kombinasyonu sayısı
    scoring='neg_mean_squared_error',    # Hedef, negatif ortalama kare hata
    cv=tscv,       # Zaman serisi çapraz doğrulama kullanılıyor
    verbose=1,     # İşlem çıktıları ekrana yazdırılıyor
    n_jobs=-1,     # Tüm CPU çekirdekleri kullanılıyor
    random_state=42     # Tekrar edilebilirlik için sabit rastgelelik
)

random_search.fit(X_train, y_train)    # Eğitim verisiyle RandomizedSearchCV çalıştırılarak en iyi parametreler bulunuyor
best_model = random_search.best_estimator_     # En iyi parametrelerle model seçiliyor

print(f"En İyi Parametreler: {random_search.best_params_}")
```

```python
# TAHMİN VE GERİ DÖNÜŞTÜRME 
pred_diffs = best_model.predict(X_test)   # Test verisi kullanılarak LightGBM modelinden fark tahminleri alınıyor

# Tahmin edilen farklar, önceki kapanış fiyatına eklenerek tahmini fiyatlar oluşturuluyor
pred_prices = original_close_test.values + pred_diffs

# Gerçek Değerleri Hizalama
final_y_true = df.loc[original_close_test.index, 'Close'].shift(-1).dropna()
final_y_pred = pd.Series(pred_prices, index=original_close_test.index).shift(1).dropna()

# Tahmin ve gerçek değerlerin ortak indeksleri belirleniyor
common_index = final_y_true.index.intersection(final_y_pred.index)
final_y_true = final_y_true.loc[common_index]
final_y_pred = final_y_pred.loc[common_index]
```

```python
# --- 5. SONUÇLAR VE PERFORMANS DEĞERLENDİRME ---

# Kök Ortalama Kare Hata (RMSE) hesaplanıyor.
rmse = np.sqrt(mean_squared_error(final_y_true, final_y_pred))

# Ortalama Mutlak Hata (MAE) hesaplanıyor.
mae = mean_absolute_error(final_y_true, final_y_pred)

# R2 skoru ile modelin açıklayıcılığı ölçülüyor
r2 = r2_score(final_y_true, final_y_pred)

print(f"\n-- LIGHTGBM + DIFFERENCING SONUÇLARI --")
print(f"RMSE (Kök Ortalama Kare Hata): {rmse:.4f}")
print(f"MAE (Ortalama Hata): {mae:.4f}")  
print(f"R2 Skoru (Başarı): {r2:.4f}")

# Grafik Çizimi
plt.figure(figsize=(15, 6))
plt.plot(final_y_true.index[-200:], final_y_true.iloc[-200:], label='Gerçek Fiyat', color='black', linewidth=2)
plt.plot(final_y_true.index[-200:], final_y_pred.iloc[-200:], label='LightGBM Tahmini', color='#FF9F1C', linestyle='--')
plt.title(f'LightGBM Performansı (MAE: {mae:.2f}$ Hata Payı ile)')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()

# Özellik Önemi
lgb.plot_importance(best_model, max_num_features=10, title="LightGBM İçin En Kritik Özellikler", color='purple')
plt.show()
```
